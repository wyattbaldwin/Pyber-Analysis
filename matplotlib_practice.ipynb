```python
%matplotlib inline

# Importing libraries

import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline


# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# plot the X and Y axis
plt.plot(x_axis, y_axis, label = 'Boston',marker = '*',color = 'Red', linewidth = 2)
#Create labels for X and Y Axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Setting limit for y axis
plt.ylim(0,45)

# setting x axis limit
# plt.xlim(0,11)

#Adding title
plt.title("PyBer Fare by Month")

#Add Legend
plt.legend()
```

```python
# Create a horizontal bar chart

# Importing libraries

import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline


# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# plot the X and Y axis
plt.barh(x_axis,y_axis, label = 'Boston',color = 'Red')
#Create labels for X and Y Axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Setting limit for y axis
plt.ylim(0,45)

# setting x axis limit
plt.xlim(0,11)

#Adding title
plt.title("PyBer Fare by Month")

#Add Legend
plt.legend()
```

```python
# Sort Y axis in ascending order
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(y_axis,x_axis,'o',color = 'red',label = 'Chicago')

plt.legend()

plt.title ("PyBer Fare by Month")

plt.xlabel("Date")

plt.ylabel("Fare ($)")

plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis,s=[i*3 for i in y_axis])
```

```python
fig,ax = plt.subplots()
ax.scatter(x_axis,y_axis,s=([i*5 for i in y_axis]),color = 'skyblue',alpha = 0.8,edgecolor = 'black',lw = 3)
```

```python
plt.pie(y_axis,labels = x_axis)
plt.show()
```

```python
explode = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,explode=explode,labels = x_axis,autopct='%1.1f%%')
plt.show()
```
